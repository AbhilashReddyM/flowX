```python
# Sample API
import mae6225
import simulation
mae6225.__version__
```

```python
# Define grid parameters
center_vars = ['asol','ivar','rvar','eror']

bc_type     = {'ivar':['dirichlet','dirichlet','dirichlet','dirichlet']}

bc_val      = {'ivar':[0.,0.,0.,0.]}

nx = 80
ny = 80

xmin = 0
xmax = 1

ymin = 0
ymax = 1

grid = mae6225.Grid(center_vars,nx,ny,xmin,xmax,ymin,ymax,bc_type,bc_val)
```

```python
# Calculate the exact solution 
simulation.get_analytical(grid,'asol')
```

```python
# Plot the exact solution
mae6225.io.plot_contour(grid,'asol')
```

```python
# Calculate the RHS
simulation.get_RHS(grid,'rvar')
```

```python
# Plot RHS
mae6225.io.plot_contour(grid,'rvar')
```

```python
# Solve Poisson equation
total_iterations,residual = mae6225.poisson.solve_jacobi(grid,'ivar','rvar',max_iterations=7000,tol=1e-9)
```

```python
# Check iterations
total_iterations
```

```python
# Check residual
residual
```

```python
# Plot the numerical solution
mae6225.io.plot_contour(grid,'ivar')
```

```python
# Calculate error
mae6225.poisson.get_error(grid,'eror','ivar','asol')
```

```python
# Plot error
mae6225.io.plot_contour(grid,'eror')
```
